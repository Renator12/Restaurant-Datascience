```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

scores=[]
for n in range(2,min(5,df_scaled.shape[0])):
  km=KMeans(n_clusters=n)
  km.fit_predict(df_scaled)
  scores.append(km.inertia_)
plt.plot([n for n in range(2,min(5,dfquant.shape[0]))],scores,marker='o')
plt.xlabel("clusters")
plt.ylabel("inertia")
```

```python
clusterdiff=np.diff(scores)
clusternum=np.argmax(np.abs(scores))+3#since our clusters start from a number of 2 we must increase it by 1since the differences start from index
km=KMeans(n_clusters=clusternum)
km.fit_predict(df_scaled)
labels=km.labels_
df['labels']=labels
clusternum
```

```python
from sklearn.preprocessing import StandardScaler

df['cluster_label'] = km.labels_
clusters=km.cluster_centers_
# Accessing cluster centers
cluster_centers = scaler.inverse_transform(km.cluster_centers_)
cluster_centers
```

```python
df_scaled[:,0]
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', c='red', s=200, label='Cluster Centers')
scat=plt.scatter(dfquant['rating'], dfquant['price'], c=df['cluster_label'], cmap='viridis', s=50)
plt.colorbar(scat, label='Cluster Label')
plt.xlabel('Distance')
plt.ylabel('Overall Rating')
plt.title('KMeans Clustering of Restaurants')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(cluster_centers[:, 1], cluster_centers[:, 2], marker='x', c='red', s=200, label='Cluster Centers')
scat=plt.scatter(dfquant['price'], dfquant['distance'], c=df['cluster_label'], cmap='viridis', s=50)
plt.colorbar(scat, label='Cluster Label')

plt.xlabel('price')
plt.ylabel('distance ')
plt.title('KMeans Clustering of Restaurants(Price vs distance)')
plt.legend()
plt.show()
```

```python
import PIL
import urllib.request
import io


def show(vals,details=False):

  for i,url1 in enumerate(vals):
    if url1 !='':
      plt.figure(figsize=(30,30))
      response = requests.get(url1)
      image_bytes = io.BytesIO(response.content)

      img = PIL.Image.open(image_bytes)
      plt.subplot(5,1,i+1)    # the number of images in the grid is 5*5 (25)
      plt.imshow(img)


      plt.title(f'{chosennames[i]}, Phone:{df[df["name"]==chosennames[i]]["phone"].values[0]} , Closed? ')
      plt.tight_layout()

  plt.show()
show(vals)
```
